```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
table = pd.read_table("fisher_ant.txt", sep='\t')
```

```python
ants = table[['ID', 'CBS4920', 'CBS134916' , 'CBS5597', 'CBS2003']]
ants['total']=ants.sum(axis=1)
sum_row_ants=ants.iloc[:, 1:29].sum(axis=0)
df_sum_ants=pd.DataFrame(data=sum_row_ants).T
df_sum_ants=df_sum_ants.reindex(columns=ants.columns)
df_final_antst=ants.append(df_sum_ants,ignore_index=True)
df_final_antst
```
